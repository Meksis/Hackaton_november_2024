```python
import requests as rq
from dotenv import load_dotenv
import dotenv

ENV_PATH = './configs.env'

load_dotenv(ENV_PATH, override=True, encoding='UTF-8')
```

```python
response = rq.get('http://localhost:3000/node/test', headers = {
    "Authorization": f"Bearer {dotenv.get_key(dotenv_path=ENV_PATH, key_to_get='AUTH_TOKEN')}"
},
params = {
    "arg1": "value1",
    "arg2": "value2"
}
)
```

```python
response.text
```
